```python
import pandas as pd 
# birank的top1000的actor和repo
birank_actor_df = pd.read_csv('./birank_actor_idname_top100.csv', index_col=0)
birank_repo_df = pd.read_csv('./birank_repo_idname_top100.csv', index_col=0)
```

```python
actor = birank_actor_df.iloc[0:50,:].copy()
actor['birank'] = actor.index+1
actor.head(5)
```

```python
burstbirank_actor_df = pd.read_csv('./burstbirank_actor_top1000.csv', index_col=0)
burstbirank_actor = burstbirank_actor_df.iloc[:50,:].copy()
burstbirank_actor['burstbirank'] = burstbirank_actor.index+1
burstbirank_actor.head(5)
```

```python
pagerank_actor_df = pd.read_csv('./pagerank_actor_top1000.csv', index_col=0)
pagerank_actor = pagerank_actor_df.iloc[:50,:].copy()
pagerank_actor['pagerank'] = pagerank_actor.index+1
pagerank_actor.head(5)
```

```python
actor_result = pd.merge(actor,burstbirank_actor, on='new_actor_id')
actor_result = pd.merge(actor_result,pagerank_actor, on='new_actor_id')
actor_result.head(10)
```

```python
repo = birank_repo_df.iloc[0:50,:].copy()
repo['birank'] = repo.index+1
repo.head(5)
burstbirank_repo_df = pd.read_csv('./burstbirank_repo_top1000.csv', index_col=0)
burstbirank_repo = burstbirank_repo_df.iloc[:50,:].copy()
burstbirank_repo['burstbirank'] = burstbirank_repo.index+1
burstbirank_repo.head(5)
pagerank_repo_df = pd.read_csv('./pagerank_repo_top1000.csv', index_col=0)
pagerank_repo = pagerank_repo_df.iloc[:50,:].copy()
pagerank_repo['pagerank'] = pagerank_repo.index+1
pagerank_repo.head(5)
repo_result = pd.merge(repo,burstbirank_repo, on='new_repo_id')
repo_result = pd.merge(repo_result,pagerank_repo, on='new_repo_id')
repo_result.head(10)
```
